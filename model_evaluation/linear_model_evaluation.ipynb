```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
```

```python
metric = pd.read_csv('linear_model_evaluation.csv')
```

```python
sample = metric[(metric['regParam'] == 0.001)&(metric['iterations'] == 1000)]
```

```python
sample = sample[2:]
```

```python
f, ax = plt.subplots(figsize=(7, 3))
plt.title('MAE with sample size increases for Linear Regression')
sns.pointplot(x=sample['SampleData'], y=sample['MAE'], hue=sample['regType'], data=sample,
              markers=["^", "o"], linestyles=["-", "--"]);
```

```python
f, ax = plt.subplots(figsize=(7, 3))
plt.title('RMSE with sample size increases for Linear Regression')
sns.pointplot(x=sample['SampleData'], y=sample['RMSE'], hue=sample['regType'], data=sample,
              markers=["^", "o"], linestyles=["-", "--"]);
```

```python
iters = metric[(metric['SampleData'] == 50)&(metric['regType'] == 'L2')&(metric['regParam'] == 0.01)]
```

```python
iters = iters[:-2]
```

```python
iters
```

```python
#f, (ax1,ax2) = plt.subplots(figsize=(7, 3))
f, (ax1, ax2) = plt.subplots(1, 2,figsize=(10, 3))
ax1.set_title('MAE with iterations increase for SGD')
sns.pointplot(x=iters['iterations'], y=iters['MAE'],  data=iters,
              markers=["^"], linestyles=["-"],ax = ax1);
ax2.set_title('RMSE with iterations increase for SGD')
sns.pointplot(x=iters['iterations'], y=iters['RMSE'],  data=iters,
              markers=["o"], linestyles=["--"],ax = ax2);
f.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9,
                wspace=0.3, hspace=0.2)
```
